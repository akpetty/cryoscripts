## Plotting/projecting/gridding -a case study with Arctic sea ice concentration
Author: Alek Petty  
Description: Simple script to show a couple different ways of projecting/gridding/plotting ice concentration data   
Input requirements: Daily ice conc data  


```python
#need this to show matplotlib figures in ipythonn notebook. 
#Switch to AGG for outputting figures
%matplotlib inline
import matplotlib
#matplotlib.use("AGG")
from mpl_toolkits.basemap import Basemap, shiftgrid
import numpy as np
from pylab import *
import numpy.ma as ma
from glob import glob
#from netCDF4 import Dataset

dataPath = '../Data/ICE_CONC/'
figpath='./Figures/'
```

```python
date='2016060'
alg=0 # NASA_TEAM
if (alg==0):
    team = 'NASA_TEAM'
    team_s = 'nt'
    header = 300
    datatype='uint8'
    scale_factor=250.
if (alg==1):
    team = 'BOOTSTRAP'
    team_s = 'bt'
    header = 0
    datatype='<i2'
    scale_factor=1000.

# use glob to find files that match that path
# can sometimes be two sensors producing data on the same day so beware!
files = glob(dataPath+'/'+team_s+'_'+date+'*.bin')

print files
```

```python
#ftp_file ='ftp://sidads.colorado.edu/pub/DATASETS/nsidc0081_nrt_nasateam_seaice/north/nt_20160612_f18_nrt_n.bin'
#fd = open(ftp_file, 'r')
fd = open(files[0], 'r')
data = fromfile(file=fd, dtype=datatype)
data = data[header:]
#FIRST 300 FILES ARE HEADER INFO
ice_conc = reshape(data, [448, 304])
#divide by scale_factor to express as concentration
ice_conc = ice_conc/scale_factor
# mask where less than some value - could also have just set these to zero
ice_conc_ma = ma.masked_where(ice_conc>1., ice_conc)

# can mask using some lower ice conc threshold
#ice_conc=ma.masked_where(ice_conc <=0.15, ice_conc)

ice_conc_ma
```

```python
# very simply plot out the raw fields on the native projection
fig = figure(figsize=(6,6))
im1 = imshow(ice_conc_ma, cmap=cm.Blues_r, vmin=0, vmax=1, zorder=2)
plt.show()
```

# Let's make this look a bit nicer...

```python
# read in the lon/lat fields that relate to the NSIDC polar sterographic projection
flat = open(dataPath+'psn25lats_v3.dat', 'rb')
flon = open(dataPath+'psn25lons_v3.dat', 'rb')
lats = fromfile(file=flat, dtype='<i4')/100000.0
lats = reshape(lats, [448, 304])
lons = fromfile(file=flon, dtype='<i4')/100000.0
lons = reshape(lons, [448, 304])

#create our projection - play around with this to see what you like!
m = Basemap(projection='npstere',boundinglat=65,lon_0=0, resolution='l'  )

xpts, ypts = m(lons, lats)
# can fill in pole hole if you want.
#ice_conc = where((lats >=85), 1, ice_conc)
```

```python
fig = figure(figsize=(6,6))
ax1=gca()

im1 = m.pcolormesh(xpts , ypts, ice_conc_ma, cmap=cm.Blues_r, vmin=0, vmax=1,shading='gouraud', zorder=2)

m.drawcoastlines(linewidth=0.25, zorder=5)
m.drawparallels(np.arange(90,-90,-10), linewidth = 0.25, zorder=3)
m.drawmeridians(np.arange(-180.,180.,30.), linewidth = 0.25, zorder=3)
m.fillcontinents(color='0.8',lake_color='0.7', zorder=2)

subplots_adjust( right = 0.99, left = 0.01, top=0.99, bottom=0.01)
#savefig(figpath+'/Arctic_BaKa_ice_extent_map_day1.png', dpi=1000)
plt.show()
```

# Or we could plot out ice extent (areas where there is some ice)

```python
ice_ext = where((ice_conc_ma >=0.15), 1, ice_conc_ma)
ice_ext = ma.masked_where((ice_ext <1), ice_ext)

fig = figure(figsize=(6,6))
ax1=gca()

im1 = m.pcolormesh(xpts , ypts, ice_ext, cmap=cm.Blues_r, vmin=0, vmax=1,shading='gouraud', zorder=2)

#set everything less than 1 to this color
m.drawmapboundary(fill_color='#587997')
m.drawcoastlines(linewidth=0.25, zorder=5)
m.drawparallels(np.arange(90,-90,-10), linewidth = 0.25, zorder=3)
m.drawmeridians(np.arange(-180.,180.,30.), linewidth = 0.25, zorder=3)
m.fillcontinents(color='0.8',lake_color='0.7', zorder=2)

subplots_adjust( right = 0.99, left = 0.01, top=0.99, bottom=0.01)
#savefig(figpath+'/Arctic_BaKa_ice_extent_map_day1.png', dpi=1000)
plt.show()
```

# Regridding data

```python
from scipy.interpolate import griddata
#create a new 100km resolution grid on this projection
# note that this fills in the pole hole. Can remove using the gridded lonG if desired.
dx_res = 200000.
nx = int((m.xmax-m.xmin)/dx_res)+1; ny = int((m.ymax-m.ymin)/dx_res)+1
grid_str=str(int(dx_res/1000))+'km'
lonsG, latsG, xptsG, yptsG = m.makegrid(nx, ny, returnxy=True)

ice_concG = griddata((xpts.flatten(), ypts.flatten()),ice_conc.flatten(), (xptsG, yptsG), method='linear')
```

```python
ice_concG_ma=ma.masked_where(np.isnan(ice_concG), ice_concG)
#ice_concG_ma=ma.masked_where(ice_concG_ma>1., ice_concG_ma)
```

```python
fig = figure(figsize=(6,6))
ax1=gca()

#use shading 'flat' i.e. not smoothed - to show the grid more clearly
im1 = m.pcolormesh(xptsG , yptsG, ice_concG_ma, cmap=cm.Blues_r, vmin=0, vmax=1,shading='flat', zorder=2)

m.drawcoastlines(linewidth=0.25, zorder=5)
m.drawparallels(np.arange(90,-90,-10), linewidth = 0.25, zorder=3)
m.drawmeridians(np.arange(-180.,180.,30.), linewidth = 0.25, zorder=3)
m.fillcontinents(color='0.8',lake_color='0.7', zorder=2)

#add if you want to zoom in on an area
#xpt, ypt = m(0, 90)
#ax1.set_xlim(0, xpt)
#ax1.set_ylim(0, ypt)

subplots_adjust( right = 0.99, left = 0.01, top=0.99, bottom=0.01)
#savefig(figpath+'/Arctic_BaKa_ice_extent_map_day1.png', dpi=1000)
plt.show()
```
